```python
import numpy as np
import pandas as pd
```

```python
user = pd.read_csv("./tuples/user.csv")
course = pd.read_csv("./tuples/course.csv")
video = pd.read_csv("./tuples/video.csv")
concept = pd.read_csv("./tuples/concept.csv")
teacher = pd.read_csv("./tuples/teacher.csv")
school = pd.read_csv("./tuples/school.csv")
paper = pd.read_csv("./tuples/paper.csv")
```

```python
print("length_all:\nuser:%d, course:%d, video:%d, concept:%d, teacher:%d, school:%d, paper:%d" %(len(user),len(course),len(video),len(concept),len(teacher),len(school),len(paper)))
```

```python
#  将文件的ID以字典的形式换成数字
def replaceID_Num(rel_data):
    id_n = {}
    count = 0
    for id in rel_data["id"]:
        if id in id_n:
            continue
        else:
            id_n[id] =count
            count += 1
    return id_n
```

```python
user_n,course_n,video_n,concept_n,teacher_n,school_n,paper_n=  replaceID_Num(user),replaceID_Num(course),replaceID_Num(video),replaceID_Num(concept),replaceID_Num(teacher),replaceID_Num(school),replaceID_Num(paper)
```

```python
user_video = pd.read_csv("./tuples/rel_user_video.csv")
user_course = pd.read_csv("./tuples/rel_user_course.csv")
video_concept = pd.read_csv("./tuples/rel_video_concept.csv")
course_concept = pd.read_csv("./tuples/rel_course_concept.csv")
course_video = pd.read_csv("./tuples/rel_course_video.csv")
school_course = pd.read_csv("./tuples/rel_school_course.csv")
school_teacher = pd.read_csv("./tuples/rel_school_teacher.csv")
teacher_course = pd.read_csv("./tuples/rel_teacher_course.csv")
concept_paper = pd.read_csv("./tuples/rel_concept_paper.csv")
```

```python
# 将目标文件的数据替换为对应的数字并写入文件
def replaceData_Num(data, new_filename, new_ID1,new_ID2):   
    columns = []
    f  = open(new_filename, "w")
    for each in data.columns:
        columns.append(each)
    for ID1,ID2 in zip(data[columns[0]],data[columns[1]]):
    #     print(user_n[ID1],video_n[ID2])
        f.write(str(new_ID1[ID1]) + ' ' + str(new_ID2[ID2]) + '\n')
    f.close()
```

```python
u_v_filename = "./KA_DATA/my_user_video.csv"
u_c_filename = "./KA_DATA/my_user_course.csv"
v_c_filename ="./KA_DATA/my_video_concept.csv"
c_c_filename = "./KA_DATA/my_course_concept.csv"
c_v_filename = "./KA_DATA/my_course_video.csv"
s_c_filename = "./KA_DATA/my_school_course.csv"
s_t_filename = "./KA_DATA/my_school_teacher.csv"
t_c_filename = "./KA_DATA/my_teacher_course.csv"
c_p_filename = "./KA_DATA/my_concept_paper.csv"

replaceData_Num(user_video, u_v_filename, user_n, video_n)
replaceData_Num(user_course, u_c_filename, user_n, course_n)
replaceData_Num(video_concept, v_c_filename, video_n, concept_n)
replaceData_Num(course_concept, c_c_filename, course_n, concept_n)
replaceData_Num(course_video, c_v_filename, course_n, video_n)
replaceData_Num(school_course, s_c_filename, school_n, course_n)
replaceData_Num(school_teacher, s_t_filename, school_n, teacher_n)
replaceData_Num(teacher_course, t_c_filename, teacher_n, course_n)
replaceData_Num(concept_paper, c_p_filename, concept_n, paper_n)
```
