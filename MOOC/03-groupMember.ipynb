```python
import numpy as np
import pandas as pd
```

```python
user_video = pd.read_csv("./KA_DATA/my_user_video.csv",header=None,sep= ' ')
```

```python
columns = ["user", "video"]
# 必须整体全部修改
user_video.columns = columns
```

```python
import random
user = [i for i in range(0,max(user_video['user'])+1)]
random.shuffle(user)
```

```python
groupID = []
groupMemberOne = []
groupMember = []
j = np.random.randint(2,5)
C_groupID = 0
C_len =0
for i in range(len(user)):
    C_len += j
    if j > 1:  
        j -= 1
        groupMemberOne.append(user[i] )
    elif j == 1:
        groupMemberOne.append(user[i] )
        groupMember.append(groupMemberOne)
        groupMemberOne =[]
        j = np.random.randint(2,5)
#         print(j)
        groupID.append(C_groupID)
        C_groupID += 1
    
if C_len > len(user) and groupMemberOne is not None:
    groupMember.append(groupMemberOne)
    C_groupID += 1
    groupID.append(C_groupID)
```

```python
filename = './KA_DATA5/moocgroupMember_u_v.csv'
GroupID = 0
count = 0
with open(filename,'w') as object:
    for i in groupMember:
        object.write(str(GroupID)+" ")
        count=0
        for j in i:
            c = len(i)
            count+=1
            if count < c:
                object.write(str(j))
                object.write(',')
            else:
                object.write(str(j))
        GroupID+=1
        object.write('\n')
```
